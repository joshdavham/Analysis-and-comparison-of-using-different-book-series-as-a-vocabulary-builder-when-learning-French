# The Simulation

## Percy Jackson

```python
#a list of all the percy jackson books we will consider for the analysis
books_list = ['Percy Jackson/1/','Percy Jackson/2/','Percy Jackson/3/']

#load the relevant book and word-frequency data for the simulation
book_dfs, freq_list_series = load_data(books_list)

#run the simulation beginning with knowledge of the most common 5000 words
vocab_size_init = 5000
learned_words_data_PJ, known_words_PJ = one_T_simulation(books_list, book_dfs, freq_list, freq_list_series, vocab_size_init)
```
